**Problem 1: Smart Parking Lot Management System**

```python
def manage_parking(capacity, vehicle_logs):
    current_parked = 0
    peak_usage = 0

    for log in vehicle_logs:
        if log == "IN":
            current_parked += 1
        elif log == "OUT" and current_parked > 0:
            current_parked -= 1

        if current_parked > peak_usage:
            peak_usage = current_parked

    print("Currently Parked Vehicles:", current_parked)

    if current_parked > capacity:
        print("Parking Status: Exceeded Capacity")
    else:
        print("Parking Status: Available")

    print("Peak Parking Usage:", peak_usage)


logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
manage_parking(50, logs)
```

**Problem 2: Online Food Delivery Time Estimator**

```python
def estimate_delivery_time(distance, traffic, weather):
    base_time = distance * 5  # 5 minutes per km

    # Traffic delay
    if traffic.lower() == "high":
        base_time += 15
    elif traffic.lower() == "medium":
        base_time += 10
    else:
        base_time += 5

    # Weather delay
    if weather.lower() == "rainy":
        base_time += 10
    elif weather.lower() == "storm":
        base_time += 20

    print("Estimated Delivery Time:", base_time, "minutes")


estimate_delivery_time(8, "High", "Rainy")
```

**Problem 3: Movie Theatre Seat Occupancy Analyzer**

```python
def analyze_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", round(occupancy), "%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")
        

booked = [1] * 150
analyze_occupancy(200, booked)
```

**Problem 4: Cloud Server Load Classification System**

```python
def classify_server_load(cpu_readings):
    total = 0

    for reading in cpu_readings:
        total += reading

    average = total / len(cpu_readings)

    print("Average CPU Load:", round(average), "%")

    if average < 50:
        print("Server Status: Normal")
    elif average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


readings = [45, 60, 70, 85, 90]
classify_server_load(readings)
```

**Problem 5: Smart Classroom Resource Usage Monitor**

```python
def monitor_resources(resource_usage):
    overused = []

    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resources(usage)
```

**Problem 6: Online Event Registration Capacity Controller**

```python
def manage_registration(capacity, registrations):
    if registrations <= capacity:
        confirmed = registrations
        waitlist = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)


manage_registration(100, 105)
```
